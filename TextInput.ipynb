```python
from datetime import datetime
def readLine(text=""):
    """
    Prints text and returns inputted answer.

    Args:
        text (str): The text that will be printed.

    Returns:
        str: Inputted answer
    """
    print(text);
    return input();
def intPrompt(text="",length=0,default="NA"):
    """
    Prints text and returns answer as an int. Loops until integer of correct length is inputted.

    Args:
        text (str): The text that will be printed.
        length (int): Length requirement of input.
        default (any): Returned value if answer requirements are not met

    Returns:
        int: Inputted number.
        any: default if answer requirements are not met

    """

    answer = readLine(text);
    if answer == '' or (len(answer) != length and length != 0):
        print("Skriv ett {}-siffrigt tal".format(length));
        return default;
    try:
        answer = int(answer)
        if answer<0:
            print("Talet måste vara positivt");
            return default;
    except ValueError:
        print("Skriv ett tal".format(length));
        return default;
    
    return answer;
def askDate(text,onlyTimeOfDay = False):
    """
    Prints text to get desired date and convert it to a datetime object. If onlyTimeOfDay is False, only time of day will be asked.

    Args:
        text (str): The text that will be printed.
        onlyTimeOfDay (bool): Will only return time of day if true.

    Returns:
        if onlyTimeOfDay == True: 
            returns datetime(YY,MM,DD,hh,mm)   
        if onlyTimeOfDay == False: 
            returns [int,int]
    """
    print(text);
    answer = "NA";
    if(onlyTimeOfDay == False):
        while(answer == "NA"):
            answer =  intPrompt("År: (Svara i format YYYY)",4)
        year = answer
        answer = "NA";
        while(answer == "NA"):
            answer =  intPrompt("Månad: (Svara i format MM)",2)
        month = answer;
        answer = "NA";
        while(answer == "NA"):
            answer =  intPrompt("Dag: (Svara i format DD)",2)
        day = answer;
        answer = "NA";
    while(answer == "NA"):
        answer =  intPrompt("Timme: (Svara i format hh)",2)
    hour = answer
    answer = "NA";
    while(answer == "NA"):
        answer =  intPrompt("Minut: (Svara i format mm)",2)
    minute = answer
    if onlyTimeOfDay == False:
        try:
            return datetime(year,month,day,hour,minute);
        except:
            print("Inkorrekt Datum, Försök igen")
            askDate(text,onlyTimeOfDay);
    else: 
        try:
            return [hour,minute];
        except:
            print("Inkorrekt Datum, Försök igen")
            askDate(text,onlyTimeOfDay);
```
